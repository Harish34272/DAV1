```python
import pandas as pd
url = 'https://statisticstimes.com/economy/world-gdp-ranking.php'
```

```python
tables = pd.read_html(url)
df = tables[1]
```

```python
df
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

# Extract GDP 2019 data and convert it to numeric values for plotting
df[('GDP (Nominal) (miillions of $)', '2019')] = pd.to_numeric(df[('GDP (Nominal) (miillions of $)', '2019')], errors='coerce')

# Sort by GDP 2019 and select top 10
top_gdp = df.sort_values(('GDP (Nominal) (miillions of $)', '2019'), ascending=False).head(10)

# Create bar chart for top 10 GDP
plt.figure(figsize=(10, 5))
plt.bar(top_gdp[('Country/Economy', 'Country/Economy')], top_gdp[('GDP (Nominal) (miillions of $)', '2019')], color='blue')
plt.xlabel('Country')
plt.ylabel('GDP (miillions of $)')
plt.title('Top 10 Countries by GDP in 2019')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a new figure
plt.figure(figsize=(10, 5))

# Get the top 10 rows of the columns
dollar_1 = df[('GDP (PPP) (miillions of Int. $)', '2019')].head(10)
dollar_2 = df[('GDP (Nominal) (miillions of $)', '2019')].head(10)

# Create an array with the position of each bar along the x-axis
x = np.arange(len(df[('Country/Economy', 'Country/Economy')].head(10)))

# Create the bar plots
plt.bar(x - 0.2, dollar_1, 0.4, label='GDP (PPP) 2019')
plt.bar(x + 0.2, dollar_2, 0.4, label='GDP (Nominal) 2019')

# Set labels and title
plt.xlabel('Country')
plt.ylabel('GDP')
plt.title('GDP nominal and PPP')

# Name the x-axis with country names
plt.xticks(x, df[('Country/Economy', 'Country/Economy')].head(10))

plt.legend()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DAV/tv.csv')
```

```python
Average_rating = df['av_rating']
Season_no = df['seasonNumber']
```

```python
# Filter the DataFrame for "Law & Order: Special Victims Unit"
df_law_order = df[df['title'] == 'Law & Order: Special Victims Unit']

# Sort the DataFrame by 'seasonNumber'
df_law_order = df_law_order.sort_values('seasonNumber')

# Create a line chart of the ratings
plt.plot(df_law_order['seasonNumber'], df_law_order['av_rating'], color='blue',marker='o')
plt.title('Historical Ratings of Law and Order Special Victims Unit')
plt.xlabel('Season Number')
plt.ylabel('Average Rating')
plt.show()
```
